## Carico il dataset fornito Placement_Data_Full_Class.train.csv

```python
import pandas as pd
dataframe = pd.read_csv("./Placement_Data_Full_Class.train.csv")
dataframe.head()
```

# Splitto i dati di test in due file csv denominati rispettivamente train.csv e test.csv

Per avere un'anteprima dei file che andr√≤ a generare mi genere un output grazie alla funzione head() 

```python
from sklearn.model_selection import train_test_split
y = dataframe['status'].copy()
X = dataframe.drop("status",axis=1).copy()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
train = X_train.join(y_train)
train.head()
```

```python
test = X_test.join(y_test)
test.head()
```

```python
train.to_csv("train.csv",index=False)
print("Salvataggio del file train.csv")
```

```python
test.to_csv("test.csv",index=False)
print("Salvataggio del fine test.csv")
```
